```python
import pandas as pd

df = pd.read_csv('data/DBData.csv')
df
```

```python
df = df.drop(['Country Code', 'Indicator Code', 'Unnamed: 20'], axis = 1)
df
```

```python
df1 = pd.DataFrame(
        df.pivot(index='Indicator Name', 
                 columns='Country Name').unstack().reset_index())
df1.head()
```

```python
df1.columns = ['Year', 'Country', 'Measure', 'Score']
df1
```

```python
df1 = df1.pivot_table(values='Score', index='Measure', columns=['Year', 'Country']
                   ).T.reset_index()
df1
```

```python
df2 = pd.read_csv('data/GDPpc.csv')
df2
```

```python
df2 = df2.drop(['Country Code', 'Indicator Code'], axis = 1)
df2
```

```python
df2 = pd.DataFrame(
        df2.pivot(index='Indicator Name', 
                 columns='Country Name').unstack().reset_index())
df2.head()
```

```python
df2.columns = ['Year', 'Country', 'Measure', 'GDP']
df2
```

```python
dfy = df2.pivot_table(values='GDP', index='Measure', columns=['Year', 'Country']
                   ).T.reset_index()
dfy
```

```python
merged = pd.merge(df1, dfy, on = ['Country', 'Year'], how = 'inner')
merged
```
